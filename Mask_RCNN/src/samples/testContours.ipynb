```python
import skimage.io
from skimage.measure import find_contours
from skimage.measure import moments
from skimage.draw import circle
import skimage.color
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

```python
img = skimage.io.imread("/home/akshay/Mask_RCNN/src/images/00.png")
gray = skimage.color.rgb2gray(img)
plt.imshow(gray)
contours = find_contours(gray,0.5)
M = moments(gray)
cx = M[1,0]/M[0,0]
cy = M[0,1]/M[0,0]
print("centroids: ", cx, cy)
fig, ax = plt.subplots()
ax.imshow(gray, interpolation='nearest', cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)
    #print ("iterating: ", contour[:,0].argmax())
    #print ("test: ", contour[0])

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
pc = Circle((cy,cx),5)
ax.add_patch(pc)
print ("displaying contour with centroid")
        
```
